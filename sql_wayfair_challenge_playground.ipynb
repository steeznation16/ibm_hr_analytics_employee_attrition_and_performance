```python
data = [[1, '08:00', '09:15'],
        [2, '13:20', '15:20'],
        [3, '10:00', '14:00'],
        [4, '13:55', '16:25'],
        [5, '14:00', '17:45'],
        [6, '14:05', '17:45']]
df = pd.DataFrame(data, columns=['ID', 'START', 'END'])
df['START'] = pd.to_datetime(df.START)
df['END'] = pd.to_datetime(df.END)
df
```

```python
df1 = df.sort_values(by=['START'])
df2 = df.sort_values(by=['START'])
print(df1)
print()
print(df2)
```

```python
df = df1.merge(df1[(df2.START > df1.START) & (df2.START < df1.END)], on='ID', how='inner')
df
```

```python
df = df.groupby('ID').count()
df
```

```python
minMeetingRooms = df.max()+1
print(minMeetingRooms)
```
